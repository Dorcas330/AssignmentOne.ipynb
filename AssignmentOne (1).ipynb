a)	Define the following terms: (3 Marks)
i. Data Structure
Data structure is a method of organizing, storing, and managing data so that it can be accessed and modified efficiently. It is the relationship between data elements and the operations that can be performed on them.

ii. Algorithm
An algorithm is a finite sequence of well-defined steps used to solve a problem or perform a computation.
iii. Abstract Data Type (ADT)
Abstract Data Type is a model of a data structure that defines the data and the operations that can be performed on it, without specifying how they are implemented.

a) Write a Python function called check_even_odd() that:
Accepts a number as a parameter
Uses a conditional statement
Returns whether the number is Even or Odd


```python
def check_even_odd(number):
    if number % 2 == 0:
        return "Even"
    else:
        return "Odd"
```
